```python
import pandas as pd
import glob
```

```python
path = r'/Users/jessicaetchechury/Desktop/JEtchCreationsDashboard/CollectedData/' # use your path
all_files = glob.glob(path + "/erank_shop_stats_df*.csv")

li = []

for filename in all_files:

    df = pd.read_csv(filename, index_col=None, header=0)
    
    df=df.pivot(index='date',
       columns='stat',
       values='value')

    li.append(df)
    
    print(filename)
    
shop_stats = pd.concat(li, axis=0, ignore_index=False)
```

```python
shop_stats
```

```python
shop_stats.reset_index(inplace=True)
shop_stats['date'] =  pd.to_datetime(shop_stats['date'], format='%m%d%y')
shop_stats
```

```python
shop_stats=shop_stats.rename(columns={'Avg. Sales/Day':'avgSales',
                                     'Category':'category',
                                     'Sales Rank (Global)':'saleRankWW',
                                     'Sales Rank (National)':'salesRankNAT',
                                     'Shop Age':'shopAge',
                                     'Total Sales':'sales'})

shop_stats.columns
```
