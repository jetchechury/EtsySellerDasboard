# ERank Shop Stats

```python
path = r'/Users/jessicaetchechury/Desktop/JEtchCreationsDashboard/CollectedData' # use your path
all_files = glob.glob(path + "/erank_shop_stats_df*.csv")

li = []

for filename in all_files:
    df0 = pd.read_csv(filename, index_col=None, header=0)
    li.append(df0)
    print(filename)
    
df1 = pd.concat(li, axis=0, ignore_index=True, sort=True)


df1.head(10)
```

```python
sales = []
record_date = []
salesRankWW = []
salesRankNAT = []
shopAge = []
avgSales = []
category = []

for index, item in df1.iterrows():
    if item[1]=="Total Sales":
        sales.append(item[2])
        record_date.append(item[0])
    if item[1]=="Sales Rank (Global)":
        salesRankWW.append(item[2])
    if item[1]=="Sales Rank (National)":
        salesRankNAT.append(item[2])
    if item[1]=="Shop Age":
        shopAge.append(item[2])
    if item[1]=="Avg. Sales/Day":
        avgSales.append(item[2])
    if item[1]=="Category":
        category.append(item[2])
        
df2 = pd.DataFrame({'record_date':record_date,
                  'avgSales':avgSales,
                  'category':category,
                  'salesRankWW':salesRankWW,
                  'salesRankNAT':salesRankNAT,
                  'shopAge':shopAge,
                  'sales':sales})

df2
```

# Competitor Sales

```python
df = pd.read_csv('/Users/jessicaetchechury/Desktop/JEtchCreationsDashboard/CollectedData/eRank_CompetitorSales_071519.csv', index_col=None, header=0)
df2 = pd.read_csv('/Users/jessicaetchechury/Desktop/JEtchCreationsDashboard/CollectedData/eRank_CompetitorSales_072219.csv', index_col=None, header=0)

li=[df,df2]

```

```python
from datetime import datetime, date, time, timedelta

li2=[]
li3=[]

today=date.today()
yesterday = date.today() - timedelta(days=1)
yesterday_year=yesterday.strftime('%Y')

for item in li:
    comp_sales_daily = item.drop(item.iloc[:,11:14],axis=1)
    comp_sales_daily.drop(comp_sales_daily.columns[[0,2,3,-1]],axis=1,inplace=True)
    # Get df column values
    comp_sales_daily_columns = comp_sales_daily.columns.get_values()

    # Turn column values to list
    comp_sales_daily_columns=comp_sales_daily_columns.tolist()
    
    id_vars=['Shop']

    # Identify values for melt
    for item in id_vars:
        comp_sales_daily_columns.remove(item)

    # Melt df 
    comp_sales_daily2=pd.melt(comp_sales_daily, id_vars=id_vars, value_vars=comp_sales_daily_columns)

    # Add year and change str to datetime
    comp_sales_daily2['date'] = yesterday_year + comp_sales_daily2['variable'].astype(str)
    comp_sales_daily2['date']= pd.to_datetime(comp_sales_daily2['date'], format="%Y%a%d-%b")

    # Drop column no longer needed 
    comp_sales_daily2=comp_sales_daily2.drop(['variable'],axis=1)
    
    # Rename value column to sales
    comp_sales_daily2=comp_sales_daily2.rename(columns={'value':'sales'})

    comp_sales_daily2['week'] = (comp_sales_daily2['date'] + pd.DateOffset(days=1)).dt.week
    comp_sales_daily2
    cur_week_no=comp_sales_daily2.iloc[0]['week']
    
    li2.append(comp_sales_daily2)
for item in li:
   
    comp_sales_sum=item.drop(item.iloc[:,4:12],axis=1)
    comp_sales_sum.drop(comp_sales_sum.columns[[0,2]],axis=1,inplace=True)
    
    # Rename original week column
    comp_sales_sum=comp_sales_sum.rename(columns={'Week':'weeklySales'})
    
    comp_sales_sum['week']=cur_week_no
    li3.append(comp_sales_sum)
```

```python
comp_sales_daily_df = pd.concat(li2, axis=0, ignore_index=True, sort=False)
comp_sales_sum_df = pd.concat(li3, axis=0, ignore_index=True, sort=False)

comp_sales_daily_df.head()
```
