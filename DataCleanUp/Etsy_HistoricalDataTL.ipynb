```python
import pandas as pd
import glob
import numpy as np
```

```python
# Concat all files with scraped data

path = r'/Users/jessicaetchechury/Desktop/JEtchCreationsDashboard/Historical_Data' # use your path
all_files = glob.glob(path + "/historical_listing_data*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
    print(filename)
    
historical_stats_df = pd.concat(li, axis=0, ignore_index=True, sort=True)

historical_stats_df.head()
```

```python
# Remove columns not needed
historical_stats_df=historical_stats_df[['listingID','visits','itemsSold',
                                            'revenue','fromDate']]

historical_stats_df.head()
```

```python
# Rename columns to match database 
historical_stats_df=historical_stats_df.rename(columns={'itemsSold':'sold','fromDate':'date'})
historical_stats_df.head()
```

```python
#Fix duplicate revenues from scraping
rev_dups=historical_stats_df.loc[historical_stats_df['revenue']>0,:]
rev_dups
```

```python
# Delete every other row
rev_dups=rev_dups.iloc[::-2]
rev_dups
```

```python
# Create database with listings who have no revenue
non_rev=historical_stats_df.loc[historical_stats_df['revenue']==0,:]
non_rev.head()
```

```python
# Concat rev and non rev info

historical_stats_df = pd.concat([rev_dups,non_rev], ignore_index=True, sort=False)
historical_stats_df
```

```python
historical_stats_df=historical_stats_df.sort_values(by=['date','listingID'],ascending=False)
historical_stats_df.head()
```
